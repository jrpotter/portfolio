## Mathematics

The following creates a dot diagram for an easier way of referencing what cards we could construct with parameters:

<pre>
S = {a, b, c, d}
k = 2
</pre>

Here we have a visual indication of the cards we could create with these given parameters. This is done by counting the number of nodes connected by a single color (in this case 3).

```haskell
{-# LANGUAGE FlexibleContexts #-}
{-# LANGUAGE QuasiQuotes #-}

import Data.String.Here
import IHaskell.Display.Graphviz as G

G.dot [here|
  digraph {
    layout=circo
    node [penwidth=1]
    edge [arrowhead="none", penwidth=2, labelfontsize=0]
    
    # label=a, color=lightcoral
    # label=b, color=palegreen4
    # label=c, color=blue
    # label=d, color=darkorange1
    
    "{a, b}" -> "{a, c}" [color=lightcoral]
    "{a, b}" -> "{a, d}" [color=lightcoral]
    "{a, b}" -> "{b, c}" [color=palegreen4]
    "{a, b}" -> "{b, d}" [color=palegreen4]
    "{a, c}" -> "{a, d}" [color=lightcoral]
    "{a, c}" -> "{b, c}" [color=blue]
    "{a, c}" -> "{c, d}" [color=blue]
    "{a, d}" -> "{b, d}" [color=darkorange1]
    "{a, d}" -> "{c, d}" [color=darkorange1]
    "{b, c}" -> "{b, d}" [color=palegreen4]
    "{b, c}" -> "{c, d}" [color=blue]
    "{b, d}" -> "{c, d}" [color=darkorange1]

    # "{a, c}" -> "{a, b}" [color=lightcoral]
    # "{a, d}" -> "{a, b}" [color=lightcoral]
    # "{a, d}" -> "{a, c}" [color=lightcoral]
    # "{b, c}" -> "{a, b}" [color=palegreen4]
    # "{b, c}" -> "{a, c}" [color=blue]
    # "{b, d}" -> "{a, b}" [color=palegreen4]
    # "{b, d}" -> "{b, c}" [color=palegreen4]
    # "{b, d}" -> "{a, d}" [color=darkorange1]
    # "{c, d}" -> "{a, c}" [color=blue]
    # "{c, d}" -> "{b, c}" [color=blue]
    # "{c, d}" -> "{a, d}" [color=darkorange1]
    # "{c, d}" -> "{b, d}" [color=darkorange1]
  }
|]
```
